```python
 A Python generator function allows you to declare a function that behaves like an iterator, providing a faster and easier way to create iterators.
      yield keyword is a special keyword which helps the user to create a generator function  
```

```python
def count_test(n):
    count=1
    while count<n :
        yield count
        count = count+1
```

```python
c = count_test(5) 
```

```python
for i in c:
    print(i) 
```

```python
def test_fib(n):
    a,b=0,1
    for i in range(n):
        yield a
        a,b = b , a+b
```

```python
for i in test_fib(10):
    print(i) 
```
